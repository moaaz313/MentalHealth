# Mental Health Project 

Mental Health Dataset from https://www.kaggle.com/datasets/bhadramohit/mental-health-dataset.

The data contains the following fields:
* User_ID: A unique number to identify each person in the data.

* Age: The age of the person.

* Gender: Whether the person is male, female, or another gender.

* Occupation: The type of job or work the person does( Sales , Education , Healthcare , IT , Engineering , Finance , Other )

* Country: The country where the person lives.
* Mental_Health_Condition: Whether the person has a mental health condition (e.g., Yes/No or specific condition names).
* Severity: How serious the mental health condition is (e.g., low, medium, high).


* Consultation_History: Information about whether the person has seen a mental health professional in the past.


* Stress_Level: How much stress the person feels (e.g., low, medium, high).

* Sleep_Hours: The number of hours the person sleeps each day.


* Work_Hours: The number of hours the person works .


* Physical_Activity_Hours: The number of hours the person spends on physical activities like exercise.

What is the relationship between gender and stress levels?

```python
# Proportions of stress levels by gender
stress_gender = pd.crosstab(df['Gender'], df['Stress_Level'], normalize='index') * 100

plt.figure(figsize=(10, 6))
stress_gender.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(10, 6))
plt.title('Stress Level Distribution by Gender (%)', fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.legend(title='Stress Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Chi-square test to check dependence
from scipy.stats import chi2_contingency
chi2, p, _, _ = chi2_contingency(pd.crosstab(df['Gender'], df['Stress_Level']))
print(f"Chi-Square Test p-value: {p:.5f}")
```
